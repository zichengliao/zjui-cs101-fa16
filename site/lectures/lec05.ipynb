```python
print( "QUADRATIC SOLVER" )
print( "a x^2 + b x + c = 0" )
a = float( input( 'a: ' ) )
b = float( input( 'b: ' ) )
c = float( input( 'c: ' ) )
root = ( b**2 - 4*a*c ) ** 0.5
denom = 2 * a
pos = (-b + root) / denom
neg = (-b - root) / denom
message1 = "%.2f + %.2fi" % (pos.real,pos.imag)
message2 = "%.2f + %.2fi" % (neg.real,neg.imag)
print("Solution 1: %s" % message1)
print("Solution 2: %s" % message2)
```
